```python
# importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv("Mall's Customers Data.csv")
```

```python
x = Data.iloc[:,[3,4]].values
```

```python
print(x)
```

```python
from sklearn.cluster import KMeans
wcss = []
```

```python
for i in range(1,21) :
    k = KMeans(n_clusters = i, init = 'k-means++', random_state=0)
    k.fit(x)
    wcss.append(k.inertia_)
```

```python
plt.plot(range(1,21), wcss)
plt.title('The elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('wcss values')
plt.show()
```

```python
# From above plot, I can fix the cluster at 5
k_model = KMeans(n_clusters = 5, init='k-means++', random_state=0)
```

```python
y_kmeans = k_model.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1], s= 100, c = 'red', label ='Customer1')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s= 100, c = 'green', label ='Customer2')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1], s= 100, c = 'blue', label ='Customer3')
plt.scatter(x[y_kmeans == 3,0], x[y_kmeans == 3,1], s= 100, c = 'yellow', label ='Customer4')
plt.scatter(x[y_kmeans == 4,0], x[y_kmeans == 4,1], s= 100, c = 'black', label ='Customer5')
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], s= 80, c= 'magenta', label='centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual income (in thosands dollars)')
plt.ylabel('Spending Score (out of 100)')
plt.legend()
plt.show()
```
